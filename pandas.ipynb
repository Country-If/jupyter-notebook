## pandas练习题

#### The previous input is as follows：

```python
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

**1. Create df with labels as index**

```python
import numpy as np
import pandas as pd
df1 = pd.DataFrame(data, index = labels)
df1
```

**2. Show df basic information and its data**

```python
df1.info()
df1.describe()
```

**3. Select only the 'animal' and 'age' columns**

```python
df3 = df1[["animal", "age"]]
df3
```

**4. Select the row with missing value**

```python
df4 = df1[df1.isnull().values==True]
df4
```

**5. Sort DF in descending age and ascending visit order**

```python
df5 = df1.sort_values(by=["age","visits"],ascending=[False,True])
df5
```

**6. In the 'animal' column, replace 'snake' with 'Python'**

```python
df6 = df1.replace("snake","Python")
df6
```
